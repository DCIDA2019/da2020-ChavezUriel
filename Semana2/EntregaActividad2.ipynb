```python
import numpy as np
import pandas as pd
```

```python
lluv1907=pd.read_csv("lluv1907.csv")
lluv1908=pd.read_csv("lluv1908.csv")
lluv1909=pd.read_csv("lluv1909.csv")
lluv1910=pd.read_csv("lluv1910.csv")
lluv1911=pd.read_csv("lluv1911.csv")
lluv1912=pd.read_csv("lluv1912.csv")

tmax09=pd.read_csv("201909010000TMax.csv")
tmax10=pd.read_csv("201910010000TMax.csv")
tmax11=pd.read_csv("201911010000TMax.csv")
tmed09=pd.read_csv("201909010000TMed.csv")
tmed10=pd.read_csv("201910010000TMed.csv")
tmed11=pd.read_csv("201911010000TMed.csv")
tmin09=pd.read_csv("201909010000TMin.csv")
tmin10=pd.read_csv("201910010000TMin.csv")
tmin11=pd.read_csv("201911010000TMin.csv")

print('Tamaños de dataframes')
print('lluv1907:', lluv1907.shape) #JULIO
print('lluv1908:', lluv1908.shape) #AGOSTO
print('lluv1909:', lluv1909.shape) # ...
print('lluv1910:', lluv1910.shape)
print('lluv1911:', lluv1911.shape)
print('lluv1912:', lluv1912.shape)

print('tmax09:', tmax09.shape) #SEPTIEMBRE
print('tmax10:', tmax10.shape) #OCTUBRE
print('tmax11:', tmax11.shape) #NOVIEMBRE
print('tmed09:', tmed09.shape)
print('tmed10:', tmed10.shape)
print('tmed11:', tmed11.shape)
print('tmin09:', tmin09.shape)
print('tmin10:', tmin10.shape)
print('tmin11:', tmin11.shape)
```

Con ayuda del comando merge se pueden unir bases de datos desechando aquellos que están fuera de la intersección. Se tienen los datos de temperaturas mínimas, medias y máximas de 3 meses, primero se crean 3 tablas, una para cada mes, uniendo con el comando pd.merge las bases de datos correspondientes a la temperatura mínima, media y máxima de cada mes. Después, estas 3 tablas se unen para formar la tabla 'merget'.

```python
mergedt09=pd.merge(pd.merge(tmin09,tmed09),tmax09)
mergedt09.index=range(1,mergedt09.shape[0]+1)
mergedt09=mergedt09.rename(columns={'Est':'Estación','Tmin':'Min09','Tmed':'Med09','Tmax':'Max09'})

mergedt10=pd.merge(pd.merge(tmin10,tmed10),tmax10)
mergedt10.index=range(1,mergedt10.shape[0]+1)
mergedt10=mergedt10.rename(columns={'Est':'Estación','Tmin':'Min10','Tmed':'Med10','Tmax':'Max10'})

mergedt11=pd.merge(pd.merge(tmin11,tmed11),tmax11)
mergedt11.index=range(1,mergedt11.shape[0]+1)
mergedt11=mergedt11.rename(columns={'Est':'Estación','Tmin':'Min11','Tmed':'Med11','Tmax':'Max11'})

mergedt=pd.merge(pd.merge(mergedt09,mergedt10),mergedt11)
mergedt
```
