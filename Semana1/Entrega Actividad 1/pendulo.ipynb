```python
import numpy as np
import matplotlib.pyplot as plt

veli=2.0  #condiciones iniciales
posi=0.0

w0=3.0  #frecuencia angular=sqrt(g/l)

it=1000  #número de iteraciones
ti=0.0  #tiempo inicial
tf=10.0  #tiempo final
h=(tf-ti)/(it+1)  #tamaño de paso

pos=np.zeros(it)  #arreglo para guardar posición
vel=np.zeros(it)  #arreglo para guardar velocidad
posreal=np.zeros(it)
velreal=np.zeros(it)

t=np.zeros(it)   #arreglo para guardar tiempo

pos[0]=posi  #aplicación de condiciones iniciales
vel[0]=veli
posreal[0]=posi
velreal[0]=veli
```

```python
for i in range(0,it-1):
    pos[i+1]=pos[i]+h*vel[i]
    vel[i+1]=vel[i]+h*(-w0**2*pos[i])
    t[i+1]=t[i]+h
    posreal[i+1]=2/w0*np.sin(w0*t[i+1])
    velreal[i+1]=2*np.cos(w0*t[i+1])
```

```python
plt.plot(t,pos,'b',label='x(t) aproximación')
plt.plot(t,posreal,'r--',label='x(t) real')
plt.xlabel('Tiempo')
plt.ylabel('Posición')
plt.title('Gráfica de posición vs tiempo')
plt.legend()

plt.savefig("Posicion.png")
```

```python
plt.plot(t,vel,'b',label='v(t) aproximación')
plt.plot(t,velreal,'r--',label='v(t) real')
plt.xlabel('Tiempo')
plt.ylabel('Velocidad')
plt.title('Gráfica de velocidad vs tiempo')
plt.legend()

plt.savefig("Velocidad.png")
```
