```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import special
from mpl_toolkits import mplot3d
```

# Actividad 1

Se obtiene una distribución normal de datos con media=70 y desviación estándar=2

```python
np.random.seed()

# Datos generados aleatoriamente, o con distribución normal
uniform=np.random.random_sample(1000)

# Datos con distribución normal obtenida con np.random.normal
data=pd.Series(np.random.normal(loc=70,scale=2,size=1000))


### PDF Probability Distribution Function
def gaussian(x):
    sigma=2
    med=70
    g=1/(sigma*(2*np.pi)**(1/2))*np.exp(-1/2*((x-med)/sigma)**2)
    return g


### CDF Cumulative Distribution Function
def cdf(x):
    sigma=2
    med=70
    arg=(x-med)/(sigma*np.sqrt(2))
    f=1/2*(1+special.erf(arg))
    return f


### INVERSE CDF
def cdfinv(x):
    sigma=2
    med=70
    f=med+sigma*np.sqrt(2)*special.erfinv(2*x-1)
    return f

a=pd.DataFrame([uniform,cdfinv(uniform)]).T.sort_values(by=[0])
```

```python
### Gráfica de pares de datos obtenidos con la función inversa de CDF
plt.figure(figsize=[6,4])
ax=plt.subplot(1,1,1)
plt.plot(a[1],a[0])
ax.set_title('Pares obtenidos con CDF')
```

```python
plt.figure(figsize=[15,12])
ax=plt.subplot(2, 2, 1)
plt.hist(a[1],bins=20)
ax.set_ylabel('Frecuencia');
ax.set_xlabel('Valor');
ax.set_title('Histograma de datos obtenidos por inversa de CDF')


ax=plt.subplot(2, 2, 2)
plt.hist(data,bins=20)
ax.set_ylabel('Frecuencia');
ax.set_xlabel('Valor');
ax.set_title('Histograma de datos obtenidos por np.random.nomral')

ax=plt.subplot(2, 2, 3)
plt.hist(a[1],bins=20,cumulative=True)
ax.set_ylabel('Frecuencia');
ax.set_xlabel('Valor');
ax.set_title('Fecuencias acumuladas de datos obtenidos por inversa de CDF')


ax=plt.subplot(2, 2, 4)
plt.hist(data,bins=20,cumulative=True)
ax.set_ylabel('Frecuencia');
ax.set_xlabel('Valor');
ax.set_title('Fecuencias acumuladas de datos obtenidos por np.random.nomral')
```

# Actividad 2
Comparación entre generación de variables uniformes. Se generaron x, y, r y theta en los intervalos, -1<x<1 , -1<y<1 , 0<r<1 y 0<theta<2*pi. Se filtraron los datos x,y tales que sqrt(x²+y²). Se graficaron estos datos en una scatter.

```python
### Generación de aleatorios
ru=np.random.random_sample(10000)
thetau=np.random.random_sample(10000)*2*np.pi
xu=(np.random.random_sample(10000)-0.5)*2
yu=(np.random.random_sample(10000)-0.5)*2

### Filtrado de las x,y tales que r<=1
r=np.sqrt(xu**2+yu**2)
xu=xu[r<=1]
yu=yu[r<=1]
```

```python
plt.figure(figsize=[16,7])
ax=plt.subplot(1,2,1)
plt.scatter(ru*np.cos(thetau),ru*np.sin(thetau),c='g',s=1)
ax.set_ylabel('y');
ax.set_xlabel('x');
ax.set_title('r y theta uniformes')

ax=plt.subplot(1,2,2)
plt.scatter(xu,yu,c='r',s=1)
ax.set_ylabel('y');
ax.set_xlabel('x');
ax.set_title('x,y,z uniformes filtrados')
```

# Actividad 3
Se realiza la misma comparación del ejercicio anterior pero en 3 dimensiones, para generar distribuciones uniformes dentro de la esfera se utilizaron dos método. Primero se generaron coordenadas x,y,z para después filtrar aquellas que el radio fuera menor o igual a 1, y el segundo método fue generando r², theta y cos(phi) uniformemente.

```python
### Generación de aleatorios
# Gráfica 1
ru=np.random.random_sample(1000)
thetau=np.random.random_sample(1000)*np.pi*2
phiu=np.random.random_sample(1000)*np.pi
# Gráfica 2
xu=(np.random.random_sample(1000)-0.5)*2
yu=(np.random.random_sample(1000)-0.5)*2
zu=(np.random.random_sample(1000)-0.5)*2
# Gráfica 3
rcuad=np.random.random_sample(1000)
cosphi=(np.random.random_sample(1000)-0.5)*2


### Filtrado de las x,y tales que r<=1
r=np.sqrt(xu**2+yu**2+zu**2)
xu=xu[r<=1]
yu=yu[r<=1]
zu=zu[r<=1]
```

```python
plt.figure(figsize=[9,8])
ax = plt.axes(projection='3d')
ax.scatter(ru*np.sin(phiu)*np.sin(thetau),ru*np.sin(phiu)*np.cos(thetau),ru*np.cos(phiu))
ax.set_title('r, theta y phi uniformes')
ax.set_ylabel('y');
ax.set_xlabel('x');
ax.set_zlabel('z');
```

```python
plt.figure(figsize=[9,8])
ax = plt.axes(projection='3d')
ax.scatter(xu,yu,zu)
ax.set_title('x,y,z uniformes filtrados')
ax.set_ylabel('y');
ax.set_xlabel('x');
ax.set_zlabel('z');
```

```python
plt.figure(figsize=[9,8])
ax = plt.axes(projection='3d')
ax.scatter(np.sqrt(rcuad)*np.sin(np.arccos(cosphi))*np.cos(thetau),np.sqrt(rcuad)*np.sin(np.arccos(cosphi))*np.sin(thetau),np.sqrt(rcuad)*cosphi)
ax.set_title('r², theta y cos(phi) uniformes')
ax.set_ylabel('y');
ax.set_xlabel('x');
ax.set_zlabel('z');
```

# Actividad 4

Se generaron datos de 0 a 1 uniformemente y se calculó su exponencial usando la variable y=exp(x). Después se hizo un histograma y una gráfica de frecuencias acumuladas para observar el tipo de distribución de probabilidad de y.

```python
xu=np.random.random_sample(10000)
y=np.exp(xu)
```

```python
plt.figure(figsize=[9,8])
ax=plt.subplot(1,1,1)
ax.scatter(xu,y,c='r')
ax.set_ylabel('y');
ax.set_xlabel('x');
ax.set_title('Pares de datos generados');
```

```python
plt.figure(figsize=[16,6])
ax=plt.subplot(1,2,1)
ax.hist(y, bins=20,color='b')
ax.set_ylabel('y');
ax.set_xlabel('x');
ax.set_title('Pares de datos generados');

ax=plt.subplot(1,2,2)
ax.hist(y, bins=20,cumulative=True,color='b')
ax.set_ylabel('y');
ax.set_xlabel('x');
ax.set_title('Pares de datos generados');
```
